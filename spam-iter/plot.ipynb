```python
import numpy as np
from utils import plot_figure
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
matplotlib.rcParams['ps.useafm'] = True
matplotlib.rcParams['pdf.use14corefonts'] = True
# matplotlib.rcParams['text.usetex'] = True

import os 
log_flag = False
color_list = sns.color_palette()
```

```python
ss = 5
eps_list =  [0, 10, 100]
path = './output/'
log_flag = True
metric = 'grd'

fig, ax = plt.subplots(figsize=(5.5, 5))

for idx, eps in enumerate(eps_list):
    plot_instance = plot_figure(algo_name=f'eps{eps}-res', sub_sample=ss, dir=path, log_flag_=log_flag, metric=metric)
    plot_instance.plot_lines(ax, color_list[idx], label='SGD-GD' + r' $\epsilon={}$'.format(eps))

ax.set_title('')
ax.set_xlabel(r"Iterations $t$", fontsize=12)
plt.legend( prop={'size': 18})
# ax.set_xscale('linear')
# ax.set_yscale('linear')
plt.xlabel(r'iterations $t$', fontsize=18)
plt.ylabel(r'$\||\nabla J(\theta_t, \theta_{t})\||^2$', fontsize=18)
ax.set_xlim([1e1, 1e5])
# ax.set_ylim([1e-7, 1e-1])
plt.tight_layout()
plt.grid(True)
plt.savefig('./fig/' + f's2-grdnorm.pdf', bbox_inches='tight')
```

```python
metric = 'train_acc'
fig, ax = plt.subplots(figsize=(5.5, 5))
insert_ax = ax.inset_axes([0.5, 0.5, 0.39,0.32])
range_insert = [1e4, 1e5, 0.6115, 0.6135]

for idx, eps in enumerate(eps_list):
    plot_instance = plot_figure(algo_name=f'eps{eps}-res', sub_sample=ss, dir=path, log_flag_=log_flag, metric=metric)
    plot_instance.plot_lines(ax, color_list[idx], label='SGD-GD' + r' $\epsilon={}$'.format(eps), shadow_flag=True,  ax_insert=insert_ax, range_=range_insert)

ax.set_xlabel(r"Iterations $t$", fontsize=12)
plt.legend( prop={'size': 18})
# ax.set_xscale('linear')
ax.set_yscale('linear')
plt.xlabel(r'iterations $t$', fontsize=18)
plt.ylabel(r'Train Accuracy', fontsize=18)
ax.set_xlim([1e2, 1e5])
ax.set_ylim([0.598, 0.614])
plt.tight_layout()
plt.grid(True)
plt.savefig('./fig/' + f's2-trainacc.pdf', bbox_inches='tight')
```

```python
metric = 'test_acc'
fig, ax = plt.subplots(figsize=(5.5, 5))
insert_ax = ax.inset_axes([0.5, 0.4, 0.39,0.32])
range_insert = [1e4, 1e5, 0.5815, 0.583]

for idx, eps in enumerate(eps_list):
    plot_instance = plot_figure(algo_name=f'eps{eps}-res', sub_sample=ss, dir=path, log_flag_=log_flag, metric=metric)
    plot_instance.plot_lines(ax, color_list[idx], label='SGD-GD' + r' $\epsilon={}$'.format(eps), shadow_flag=True, ax_insert=insert_ax, range_=range_insert)

ax.get_legend().remove()
ax.set_xlabel(r"Iterations $t$", fontsize=12)
plt.legend( prop={'size': 18})
plt.xlabel(r'iterations $t$', fontsize=18)
plt.ylabel(r'Test Accuracy', fontsize=18)
plt.tight_layout()
plt.grid(True)

ax.set_xlim([1e2, 1e5])
ax.set_ylim([0.5675, 0.584])
ax.set_yscale('linear')

plt.savefig('./fig/' + f's2-testacc.pdf', bbox_inches='tight')
```
