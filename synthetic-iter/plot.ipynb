```python
import numpy as np
from util import *
from helper_plot import plot_figure
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
matplotlib.rcParams['ps.useafm'] = True
matplotlib.rcParams['pdf.use14corefonts'] = True

import os 
log_flag = False
color_list = sns.color_palette()
```

```python
fig, ax = plt.subplots(figsize=(5.5, 5))

ss = 1
eps_list =  [0, 0.1, 0.5, 2]
path = './output-cnt/' 
log_flag = True
metric = 'loss'
insert_ax = ax.inset_axes([0.57, 0.17, 0.38,0.32])
range_insert = [1e4, 1e6]

for idx, eps in enumerate(eps_list):
    plot_instance = plot_figure(algo_name=f'eps{eps}-sgd-res', sub_sample=ss, dir=path, xdata='iter', ydata=metric)
    plot_instance.plot_lines(ax, color_list[idx], label='SGD-GD' + r' $\epsilon={}$'.format(eps), ax_insert=insert_ax, range_=range_insert)


plt.legend( loc=1, prop={'size': 20})
ax.set_yscale('linear')
plt.xlabel(r'Iterations $t$', fontsize=20)
plt.ylabel(r'$V(\theta_{t})$ ', fontsize=20)
ax.set_xlim([1e1, 1e5])
plt.tight_layout()
plt.grid(True)
plt.savefig('figs/s1-loss.pdf', format='pdf')
```

```python
fig, ax = plt.subplots(figsize=(5.5, 5))

ss = 5
eps_list = [0, 0.1, 0.5, 2]
log_flag = True
metric = 'grd'


for idx, eps in enumerate(eps_list):
    plot_instance = plot_figure(algo_name=f'eps{eps}-sgd-res', sub_sample=ss, dir=path, xdata='iter', ydata=metric)
    plot_instance.plot_lines(ax, color_list[idx], label='SGD-GD' + r' $\epsilon={}$'.format(eps), shadow_flag=True, legend=True)

ax.set_xlabel(r"Iterations $t$", fontsize=22)
ax.legend( prop={'size': 18})
ax.set_xscale('log')
ax.set_yscale('log')
plt.ylabel(r'$\||\nabla J(\theta_t, \theta_{t})\||^2$', fontsize=22)
ax.set_xlim([1e2, 0.2*1e6])
plt.tight_layout()
plt.grid(True)
plt.savefig('figs/s1-grdnorm.pdf', format='pdf')
```
