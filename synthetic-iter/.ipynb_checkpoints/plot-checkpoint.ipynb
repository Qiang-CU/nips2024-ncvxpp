```python
import numpy as np
from util import plot_figure
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
matplotlib.rcParams['ps.useafm'] = True
matplotlib.rcParams['pdf.use14corefonts'] = True
# matplotlib.rcParams['text.usetex'] = True

import os 
log_flag = False
color_list = sns.color_palette()
```

### regularizer $\beta=1e-3$

```python
fig, ax = plt.subplots(figsize=(5, 4))

ss = 1
eps_list = [0, 0.1, 0.5, 2] 
path = './output-c1/' #output-c1 is correct data
log_flag = True
metric = 'loss'

for idx, eps in enumerate(eps_list):
    plot_instance = plot_figure(algo_name=f'eps{eps}-res', sub_sample=ss, dir=path, log_flag_=log_flag, metric=metric)
    plot_instance.plot_lines(ax, color_list[idx], label='SGD-GD' + r' $\epsilon={}$'.format(eps))

ax.set_xlabel(r"Iterations $t$", fontsize=12)
plt.legend( loc=1, prop={'size': 14})
# ax.set_xscale('linear')
# ax.set_yscale('linear')
plt.xlabel(r'iterations $t$', fontsize=18)
plt.ylabel(r'loss ', fontsize=18)
ax.set_xlim([1e3, 5*1e5])
# ax.set_ylim([1e-1, 1e2])
# plt.xticks([5000, 10000, 15000, 20000]) 
# plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
plt.tight_layout()
plt.grid(True)
plt.savefig('figs/s1-loss.pdf', format='pdf')
```

```python
fig, ax = plt.subplots(figsize=(5, 4))

path = './output/' #output-c1 is correct data
ss = 5
eps_list = [0, 0.1, 0.5, 2, 10]
log_flag = True
metric = 'grd'

for idx, eps in enumerate(eps_list):
    plot_instance = plot_figure(algo_name=f'eps{eps}-gd-res', sub_sample=ss, dir=path, log_flag_=log_flag, metric=metric)
    plot_instance.plot_lines(ax, color_list[idx], label='SGD-GD' + r' $\epsilon={}$'.format(eps), shadow_flag=True)

# path = './output-fullbatch'
# for idx, eps in enumerate(eps_list):
#     plot_instance = plot_figure(algo_name=f'eps{eps}-res', sub_sample=ss, dir=path, log_flag_=log_flag, metric=metric)
#     plot_instance.plot_lines(ax, color_list[idx], label='SGD-GD' + r' $\epsilon={}$'.format(eps), shadow_flag=True, line='--', legend=False)


ax.set_xlabel(r"Iterations $t$", fontsize=12)
# plt.legend( prop={'size': 12})
ax.set_xscale('log')
ax.set_yscale('log')
plt.xlabel(r'iterations $t$', fontsize=18)
plt.ylabel(r'$\|\nabla J(\theta_t, \theta_{t})\|^2$', fontsize=18)
# ax.set_xlim([3*1e2, 0.7*1e6])
# ax.set_ylim([1e-5, 1e-2])
plt.tight_layout()
plt.grid(True)
plt.savefig('figs/s1-grdnorm.pdf', format='pdf')
```

```python
fig, ax = plt.subplots(figsize=(5, 4))

ss = 5
log_flag = True
metric = 'train_acc'
path = './output-c1/'

for idx, eps in enumerate(eps_list):
    plot_instance = plot_figure(algo_name=f'eps{eps}-res', sub_sample=ss, dir=path, log_flag_=log_flag, metric=metric)
    plot_instance.plot_lines(ax, color_list[idx], label='SGD-GD' + r' $\epsilon={}$'.format(eps))

ax.set_xlabel(r"Iterations $t$", fontsize=12)
plt.legend( prop={'size': 12})
ax.set_xscale('log')
ax.set_yscale('linear')
plt.xlabel(r'iterations $t$', fontsize=18)
plt.ylabel(r'Train Accuracy', fontsize=18)
ax.set_xlim([1e3, 0.8*1e6])
# ax.set_ylim([1e-1, 1e2])
plt.tight_layout()
plt.grid(True)
plt.savefig('figs/s1-trainacc.pdf', format='pdf')
```

```python
fig, ax = plt.subplots(figsize=(5, 4))

ss = 5
log_flag = True
metric = 'test_acc'

for idx, eps in enumerate(eps_list):
    plot_instance = plot_figure(algo_name=f'eps{eps}-res', sub_sample=ss, dir=path, log_flag_=log_flag, metric=metric)
    plot_instance.plot_lines(ax, color_list[idx], label='SGD-GD' + r' $\epsilon={}$'.format(eps))

ax.set_xlabel(r"Iterations $t$", fontsize=12)
# ax.get_legend().remove()
ax.set_xscale('log')
ax.set_yscale('linear')
plt.xlabel(r'iterations $t$', fontsize=18)
plt.ylabel(r'Test Accuracy', fontsize=18)
ax.set_xlim([1e3, 0.6*1e6])
# ax.set_ylim([1e-1, 1e2])
plt.tight_layout()
plt.grid(True)
plt.savefig('figs/s1-testacc.pdf', format='pdf')
```
