```python
import numpy as np
from utils import plot_figure
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
matplotlib.rcParams['ps.useafm'] = True
matplotlib.rcParams['pdf.use14corefonts'] = True
# matplotlib.rcParams['text.usetex'] = True

import os 
log_flag = False
color_list = sns.color_palette()
```

```python
ss = 1
eps = 10
batch = 4
K_list = [4]
path = './output/'
log_flag = True
metric = 'grd'

fig, ax = plt.subplots(figsize=(5.5, 5))

for idx, K in enumerate(K_list):
    plot_instance = plot_figure(algo_name=f'eps{eps}-batch{batch}-K{K}-res', sub_sample=ss, dir=path, log_flag_=log_flag, metric=metric)
    plot_instance.plot_lines(ax, color_list[idx], label='Lazy' + r' $K={}, b={}$'.format(K, batch), smoothing=[True, 10], annotate=[True, 1e3, 2*1e-6])

# lazy 
batch=1
K=16

plot_instance = plot_figure(algo_name=f'eps{eps}-batch{batch}-K{K}-res', sub_sample=ss, dir=path, log_flag_=log_flag, metric=metric)
plot_instance.plot_lines(ax, color_list[2], label='Lazy' + r' $K={}, b={}$'.format(K, batch), smoothing=[True, 10], annotate=[True, 1e3, 1.1* 1e-5])

# Greedy Deployment
K = 1
batch=16
plot_instance = plot_figure(algo_name=f'eps{eps}-batch{batch}-K{K}-res', sub_sample=ss, dir=path, log_flag_=log_flag, metric=metric)
plot_instance.plot_lines(ax, color_list[3], label='Greedy' + r' $K={}, b={}$'.format(K, batch), smoothing=[True, 10], annotate=[True, 1e3, 0.6*1e-6])

ax.set_title(rf'$\epsilon={eps}$')
ax.set_xlabel(r"Iterations $t$", fontsize=12)
plt.legend(loc=1, prop={'size': 15})
# ax.legend([])
# ax.set_xscale('linear')
# ax.set_yscale('linear')
plt.xlabel(r'no of sample accessed', fontsize=18)
plt.ylabel(r'$\||\nabla J(\theta_t, \theta_{t})\||^2$', fontsize=18)
ax.set_xlim([1e2, 1e6])
ax.set_ylim([1e-7, 1e-1])
plt.tight_layout()
plt.grid(True)
# plt.savefig('./fig/' + f's2-lz-grdnorm-eps{eps}.pdf', bbox_inches='tight')
```
