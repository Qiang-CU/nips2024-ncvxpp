```python
import numpy as np
from util import plot_figure
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
matplotlib.rcParams['ps.useafm'] = True
matplotlib.rcParams['pdf.use14corefonts'] = True
# matplotlib.rcParams['text.usetex'] = True

import os 
log_flag = False
color_list = sns.color_palette()
```

```python
fig, ax = plt.subplots(figsize=(5.5, 5))

ss = 2
eps = 2

path = './output/' #output-c1 is correct data
log_flag = True
metric = 'loss'
### 内嵌图
inset_ax = ax.inset_axes([0.63, 0.2, 0.35,0.32])
range_insert = [5*1e5, 1e6, 0.355, 0.35475] # left right up down


# greedy deployment
K = 1
batch = 25
plot_instance = plot_figure(algo_name=f'eps{eps}-sgd-batch{batch}-K{K}-res', sub_sample=ss, dir=path, log_flag_=log_flag, metric=metric)
plot_instance.plot_lines(ax, color_list[0], label='Greedy' + r' $K={}, b={}$'.format(K, batch), ax_insert =inset_ax, range_ = range_insert, legend=True)

K = 1
batch = 5
plot_instance = plot_figure(algo_name=f'eps{eps}-sgd-batch{batch}-K{K}-res', sub_sample=ss, dir=path, log_flag_=log_flag, metric=metric)
plot_instance.plot_lines(ax, color_list[1], label='Greedy' + r' $K={}, b={}$'.format(K, batch), ax_insert =inset_ax, range_ = range_insert, legend=True)

# lazy deployment 
K = 5
batch =5
plot_instance = plot_figure(algo_name=f'eps{eps}-sgd-batch{batch}-K{K}-res', sub_sample=ss, dir=path, log_flag_=log_flag, metric=metric)
plot_instance.plot_lines(ax, color_list[2], label='Lazy' + r' $K={}, b={}$'.format(K, batch), ax_insert =inset_ax, range_ = range_insert, legend=True)

# other lazy deployment
batch = 5
K_list = [10]
for K in K_list:
    plot_instance = plot_figure(algo_name=f'eps{eps}-sgd-batch{batch}-K{K}-res', sub_sample=ss, dir=path, log_flag_=log_flag, metric=metric)
    plot_instance.plot_lines(ax, color_list[3], label='Lazy' + r' $K={}, b={}$'.format(K, batch), ax_insert =inset_ax, range_ = range_insert, legend=True)


ax.set_xlabel(r"no of sample accessed", fontsize=22)
ax.set_ylabel(r'loss ', fontsize=22)
plt.legend( loc=1, prop={'size': 15})
ax.set_yscale('linear')
ax.set_xlim([0.5*1e2, 1e6])
plt.tight_layout()
plt.grid(True)
plt.savefig('figs/s1-lz-loss.pdf', format='pdf')
```
